```python
def csv_downloader(URL:str, name:str, path:str):
    df = pd.read_csv(URL)
    df.to_csv(f'{path}/{name}.csv', index = False)
    print(f'The {name}.csv was successfully stored in {path}')
    return df
```

```python
df_orders = csv_downloader(URL = 'https://raw.githubusercontent.com/hovhannisyan91/data_analytics_with_python/refs/heads/main/data/rfm/orders.csv',
               name = 'orders',
               path = '../data')
```

```python
df_orders = pd.read_csv('../data/orders.csv', parse_dates = ['order_date'])
```

```python
fig = px.histogram(
    data_frame = df_orders,
    x = 'revenue',
    nbins = 30,
    title = 'Histogram of Revenue Distribution'
)

fig.update_layout(
    xaxis_title='Revenue',
    yaxis_title='Count'
)

fig.show()
```
