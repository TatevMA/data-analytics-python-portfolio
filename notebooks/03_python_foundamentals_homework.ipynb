```python
'''Part 3
Segmentation rules:

“Premium” if revenue > 250
“Standard” if 100 < revenue ≤ 250
“Low” otherwise
Create a list of segment labels.
Count how many customers fall into each segment.
Add Markdown explanation: Why does the order of elif statements matter?'''
```

```python
sales = [120, 250, 80, 310, 95]
segments = []

for value in sales:
    if value > 250:
        segments.append("Premium"),
    elif 100 < value < 250:
        segments.append("Standard"),
    else:
        segments.append("Low")
print("Segments:", segments)
```

```python
premium_count = 0
standard_count = 0
low_count = 0

for item in segments:
    if item == "Premium":
        premium_count += 1
    elif item == "Standard":
        standard_count += 1
    else:
        low_count += 1
print("Premium:", premium_count)
print("Standard:", standard_count)
print("Low:", low_count)
```

```python
# Add Markdown explanation: Why does the order of elif statements matter?
'''The order of elif statements is important because python checks conditions from top to bottom and as far as the condition is true
it executes it without considering the next condition.'''
```

```python
'''Part 6
Using:

revenues = [120, 250, 80, 310, 95]

Create a list of revenues after 10% tax.
Create a list of revenues greater than 100.
Create a list labeling each revenue as:
“High” if > 200
“Normal” otherwise
Compare list comprehension vs loop in Markdown.'''
```

```python
revenues = [120, 250, 80, 310, 95]
revenues_after_tax = [value * 0.9 for value in revenues]
revenues_after_tax
```

```python
revenues_greater_10 = [value for value in revenues if value > 100]
revenues_greater_10
```

```python
revenues = [120, 250, 80, 310, 95]

labels = ["High" if value > 200 else "Normal" for value in revenues]
print("Labels:", labels)

```

```python
'''| List Comprehension           | Loop                            |
| ------------------------------- | ------------------------------- |
| Shorter and cleaner             | Longer but easier for beginners |
| Good for simple transformations | Better for complex logic        |
| Often faster                    | More flexible                   |
| Creates lists in one line       | Uses multiple lines             |

List comprehensions are useful when you want to create a new list in a concise and readable way.'''
```
