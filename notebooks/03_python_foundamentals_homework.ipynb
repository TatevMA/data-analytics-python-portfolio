```python
sales = [120, 250, 80, 310, 95]
total_revenue = 0

for value in sales:
    total_revenue += value

print("Total revenue:", total_revenue)
```

```python
average_revenue = total_revenue / len(sales)
print("Average revenue:", average_revenue)
```

```python
# Identify the maximum value without using max()
sales = [120, 250, 80, 310, 95]
maximum_amount = sales[0]

for value in sales:
    if value > maximum_amount:
        maximum_amount = value

print("Maximum amount is:", maximum_amount)
```

```python
# Count how many values are greater than 150.
sales = [120, 250, 80, 310, 95]
greater_than_150 = 0
for value in sales:
    if value > 150:
        greater_than_150 += 1

print(greater_than_150, 'values are greater than 150')
```

```python
'''Part 3
Segmentation rules:

“Premium” if revenue > 250
“Standard” if 100 < revenue ≤ 250
“Low” otherwise
Create a list of segment labels.
Count how many customers fall into each segment.
Add Markdown explanation: Why does the order of elif statements matter?'''
```

```python
sales = [120, 250, 80, 310, 95]
segments = []

for value in sales:
    if value > 250:
        segments.append("Premium"),
    elif 100 < value < 250:
        segments.append("Standard"),
    else:
        segments.append("Low")
print("Segments:", segments)
```

```python
premium_count = 0
standard_count = 0
low_count = 0
for item in segments:
    if item == "Premium":
        premium_count += 1
    elif item == "Standard":
        standard_count += 1
    else:
        low_count += 1
print("Premium:", premium_count)
print("Standard:", standard_count)
print("Low:", low_count)
```

```python
# Add Markdown explanation: Why does the order of elif statements matter?
#The order of elif statements is important because python checks conditions from top to bottom and once the condition is true it executes it.
```
