## Creating a DataFrame

```python
import pandas as pd

data = {
    "name": ["Anna", "David", "Mariam"],
    "revenue": [150, 220, 90],
    "city": ["Yerevan", "Tbilisi", "Warsaw"]
}

df = pd.DataFrame(data)
df
```

```python
df.info()
df.shape
df.columns
```

# Basic DataFrame Manipulation

## Selecting Columns

```python
df["revenue"]
df[["name", "revenue"]]
```

## Adding a Column

```python
t = 0.2
df["revenue_ater_tax"] = df["revenue"] * (1 - t)
df
```

## Removing a Column

```python
df.drop("revenue_after_tax", axis = 1)
```

## Filtering Rows

```python
df[df["revenue"] > 100]
```

```python
df
```

## Updating Values

```python
df.loc[df["revenue"] < 100, "segment"] = "Low"
df.loc[df["revenue"] > 100, "segment"] = "High"
df
```

# Conditional Statements

## The Basic if Statement

```python
revenue = 150
if revenue > 100:
    print("High Revenue"),
print("Analysis Complete")
```

```python
revenue = 180
if revenue > 200:
    print("Very High revenue"),
elif revenue > 100:
    print("High revenue"),
elif revenue > 50:
    print("Medium revenue"),
else:
    print("Low revenue")
```

## Indentation | Why It Matters

```python
revenue = 150

if revenue > 100:
    print("High revenue")

print("Analysis complete")
```

## Using elif for Multiple Conditions

```python
revenue = 280

if revenue > 200:
    print("Very high revenue")
elif revenue > 100:
    print("High revenue")
elif revenue > 50:
    print("Medium revenue")
else:
    print("Low revenue")
```

## Boolean Expressions and Comparison Operators

```python
100 == 50
100 > 50
100 != 50
```

## Combining Conditions with Logical Operators

```python
revenue = 150
is_active = True

if revenue > 100 and is_active:
    print("Target customer")
```

```python
is_active = False

if not is_active:
    print("Customer is inactive")
```

## Nested Conditional Statements

```python
revenue = 150
region = "EU"

if revenue > 100:
    if region == "EU":
        print("High EU revenue")
    else:
        print("High non-EU revenue")
```

# Loops

## The Basic for Loop

```python
sales = [100, 200, 150]

for value in sales:
    print(value)
```

## Loop With Accumulation

```python
sales = [100, 200, 150]

total = 0

for value in sales:
    total = total + value

total
```

```python
sales = [100, 200, 150]
sum = sum(sales)
sum
```

## Loop With Conditional Logic

```python
sales = [100, 200, 150, 50]

for value in sales:
    if value > 100:
        print("High value:", value),
    else:
        print("Low value:", value)
```

## Looping Over Dictionaries

```python
customer = {
    "name": "Anna",
    "revenue": 150,
    "city": "Yerevan"
}

for key in customer:
    print(key, ":", customer[key])
```

```python
customer.items()
```

```python
customer = {
    "name": "Anna",
    "revenue": 150,
    "city": "Yerevan"
}

for key in customer:
    print(key, ":", customer[key])
```

```python
for key, value in customer.items():
    print(key, value)
```

## The range() Function

## Nested Loops

```python
for i in range(3):
    for j in range(2):
        print("i =", i, ", j =", j)
```

## List Comprehension

```python
# traditional loop:
sales = [100, 200, 150]

new_sales = []

for value in sales:
    new_sales.append(value * 1.2)

new_sales
```

```python
# Now using list comprehension:
sales = [100, 200, 150]

new_sales = [value * 1.2 for value in sales]
new_sales
```

## With Conditional Filtering

```python
# You can add a condition:
sales = [100, 200, 150, 50]

high_sales = [value for value in sales if value > 120]
high_sales
```

```python
# Traditional version:
high_sales = []

for value in sales:
    if value > 120:
        high_sales.append(value)

high_sales
```
